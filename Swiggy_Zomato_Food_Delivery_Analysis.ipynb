```python
#Swiggy/Zomato Food Delivery Analysis
import statistics
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Analysis delivery data
order_time=[12,1,2,11,1]
Delivery_time=[1,2,3,1,3]
deliver_partner=[1,3,3,1,2]
cust_rating=[5,4,5,3,4]
data=pd.DataFrame({
    "order_time":order_time,
    "delivery_time":Delivery_time,
    "deliver_partner":deliver_partner,
    "cust_rating":cust_rating
})
print(data)
print("order time",order_time)
print("delivery time",Delivery_time)
print("deliver partner",deliver_partner)
print("customer rating",cust_rating)
```

```python
#finding average delivery time
avg_del_time=statistics.mean(Delivery_time)
print("average delivery time",avg_del_time)
```

```python
#peak order hours
peak_order_hours=statistics.mode(order_time)
print("peak order hours",peak_order_hours)
```

```python
#most delayed area
most_delayed_area=statistics.mode(Delivery_time)
print("most delayed area",most_delayed_area)
```

```python
#top rating
data = {'Partner': deliver_partner, 'Rating': cust_rating}
df = pd.DataFrame(data)
top_partner = df.groupby('Partner')['Rating'].mean().idxmax()
top_rating = df.groupby('Partner')['Rating'].mean().max()

print(f"Top Partner: {top_partner} with a {top_rating:.1f} average rating")
```

```python
# Get the count of orders for each delivery partner
partner_counts = df['Partner'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(partner_counts,labels=partner_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Orders by Delivery Partner')
plt.axis('equal')
plt.show()
```

```python
#most delayed area time
data['order_hour'] = data['order_time'].astype(int)
avg_by_hour = data.groupby('order_hour')['delivery_time'].mean().sort_values(ascending=False)
print("\nAverage delivery time by order hour:")
print(avg_by_hour)
```
