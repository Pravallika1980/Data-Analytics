```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics

df=pd.read_csv("/content/archive (1).zip")
df.info()
#averga temp
#total temp
#total temp increased in a day
```

```python
df.describe()
```

```python
#averga temp
avg_temp=df['Temperature'].mean()
print(avg_temp)
```

```python
#total temp
total_temp=df['Temperature'].sum()
print(total_temp)
```

```python
#total temp increased in a day
total_temp_increased=df['Temperature'][1]-df['Temperature'][2]
print(total_temp_increased)
```

```python
#Prediction
prediction=total_temp_increased+avg_temp
print(prediction)

if prediction<20:
  # print(rainfall)
  print("It might rain!")
else:
  # print(no_rainfall)
  print("It might not rain!")
```

```python
graph=df.groupby("Rain")["Temperature"].mean()
graph.plot(kind="bar",color="royalblue")
plt.xlabel("weather")
plt.ylabel("Average Temperature")

plt.show()
```
