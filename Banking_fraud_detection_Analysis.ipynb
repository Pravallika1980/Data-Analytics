```python
#banking Fraud detection
import statistics
import matplotlib.pyplot as plt
import pandas as pd
#analysis data
import statistics

transaction_amount = [100, 2000, 150, 5000, 300, 7000]
location = ["IND", "IND", "US", "IND", "UK", "US"]
time = [10, 2, 14, 3, 18, 1]
account_activity = [5, 20, 3, 25, 4, 30]

# 1. Unusual Transactions (greater than average)
avg_amount = statistics.mean(transaction_amount)

unusual = []
for i in range(len(transaction_amount)):
    if transaction_amount[i] > avg_amount:
        unusual.append(i)

print("Unusual Transactions Index:", unusual)


# 2. High Risk Accounts
high_risk = []
for i in range(len(transaction_amount)):
    if transaction_amount[i] > avg_amount and account_activity[i] > statistics.mean(account_activity):
        high_risk.append(i)

print("High Risk Accounts:", high_risk)


# 3. Fraud Patterns
fraud = []
for i in range(len(time)):
    if time[i] < 6 and transaction_amount[i] > avg_amount:
        fraud.append(i)

print("Fraud Pattern Index:", fraud)
```
