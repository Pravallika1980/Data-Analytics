```python
# Hospital Patient Monitoring System

import pandas as pd
import matplotlib.pyplot as plt
import statistics

# Patient Data
name = ["anand", "padma", "Praveen", "ramnat", "jyothikka", "vinodini"]
ages = [70, 89, 65, 81, 60, 20]
condition = ["Critical", "Critical", "Good", "critical", "Good", "Normal"]
oxygen = [92, 88, 97, 90, 95, 99]
avail_beds = 3

print("Names:", name)
print("Ages:", ages)
print("Condition:", condition)
print("Oxygen Levels:", oxygen)
print("Available Beds:", avail_beds)

# Priority List
priority = []

print("\n--- Patient Prescription Details ---")

for i in range(len(name)):

    # Critical Patients
    if condition[i].lower() == "critical":

        priority.append("High Priority")

        print(f"\n{name[i]} needs immediate doctor checkup")

        # Bed Allocation
        if avail_beds > 0:
            print("Bed Allocated")
            avail_beds -= 1
        else:
            print("No Beds Available")

        # Oxygen Check
        if oxygen[i] < 90:
            print("Provide Oxygen Support")
        else:
            print("Monitor Oxygen Regularly")

        # Age-Based Care
        if ages[i] > 75:
            print("Senior Citizen Care Required")
        else:
            print("Regular ICU Observation")

    # Good Condition Patients
    elif condition[i].lower() == "good":

        priority.append("Medium Priority")

        print(f"\n{name[i]} is stable")

        if oxygen[i] < 95:
            print("Give Mild Oxygen Support")
        else:
            print("Normal Observation")

        if ages[i] > 60:
            print("Daily Health Monitoring")
        else:
            print("General Ward Observation")

    # Normal Patients
    else:

        priority.append("Low Priority")

        print(f"\n{name[i]} is normal")

        if oxygen[i] >= 98:
            print("Healthy Oxygen Level")

        print("Home Care Suggested")

# Create DataFrame
df = pd.DataFrame({
    "Name": name,
    "Age": ages,
    "Condition": condition,
    "Oxygen": oxygen,
    "Priority": priority
})

print("\n--- Hospital Patient Data ---")
print(df)

# Average Oxygen Level
avg_oxygen = statistics.mean(oxygen)
print("\nAverage Oxygen Level:", avg_oxygen)

# Save Data to CSV
df.to_csv("hospital_patient_data.csv", index=False)
print("\nDataset saved as hospital_patient_data.csv")

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(name, oxygen)
plt.title("Patient Oxygen Levels")
plt.xlabel("Patient Names")
plt.ylabel("Oxygen Level")
plt.xticks(rotation=45)
plt.show()
```
