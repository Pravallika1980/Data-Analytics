```
import pandas as pd
import numpy as np

# Data
cus_names = ["Akhi", "Asri", "anu", "ammu", "smiley"]
internet_speed = [1, 5, 10, 20, 15]
payment_failure = [1, 0, 2, 3, 2]
delivery_charges = [42, 34, 52, 67, 20]
purchase_completed = [True, False, False, True, True]

# Create DataFrame
df = pd.DataFrame({
    "Customer": cus_names,
    "Internet_Speed": internet_speed,
    "Payment_Failure": payment_failure,
    "Delivery_Charges": delivery_charges,
    "Purchase_Completed": purchase_completed
})

# ---------------- INTERNET DIAGNOSIS ----------------
avg_speed = np.mean(internet_speed)

if avg_speed < 5:
    internet_diagnosis = "Poor Internet"
elif avg_speed < 15:
    internet_diagnosis = "Moderate Internet"
else:
    internet_diagnosis = "Good Internet"

# ---------------- PAYMENT DIAGNOSIS ----------------
total_payment_failures = sum(payment_failure)

if total_payment_failures > 5:
    payment_diagnosis = "High Payment Failure"
else:
    payment_diagnosis = "Payment System Stable"

# ---------------- DELIVERY DIAGNOSIS ----------------
avg_delivery = np.mean(delivery_charges)

if avg_delivery > 50:
    delivery_diagnosis = "Delivery Charges Too High"
else:
    delivery_diagnosis = "Delivery Charges Acceptable"

# ---------------- ROOT CAUSE ANALYSIS ----------------
if total_payment_failures > 5:
    root_cause = "Payment Gateway Issues"
elif avg_speed < 5:
    root_cause = "Slow Internet"
elif avg_delivery > 50:
    root_cause = "High Delivery Charges"
else:
    root_cause = "No Major Issues"

# ---------------- OUTPUT ----------------
print("Internet Diagnosis:", internet_diagnosis)
print("Payment Diagnosis:", payment_diagnosis)
print("Delivery Diagnosis:", delivery_diagnosis)
print("Root Cause:", root_cause)
```

```
import statistics
import matplotlib.pyplot as plt
Cust_names=["pravi","mouni","manvi","nikki","vinu"]

internet_speed=[2,4,4,5,3]
payment_failure=[0,1,1,1,0,]
delivery_charges=[50,100,70,40,80]
purchase_completed=["false","true","true","true","false"]
print("customer names=",Cust_names)
print("internet speed=",internet_speed)
print("payment failure=",payment_failure)
print("delivery charges=",delivery_charges)
print("purchase completed=",purchase_completed)
```

```
for i in range(len(Cust_names)):
    print(Cust_names[i],internet_speed[i],payment_failure[i],delivery_charges[i],purchase_completed[i])
```

```
total_abandoned=purchase_completed.count("false")
print("total abandoned=",total_abandoned)
```

```
#internet diagnosis
print('\n--- Internet Speed Diagnosis ---')
for i in range(len(Cust_names)):
  customer = Cust_names[i]
  internet_spd = internet_speed[i]
  purchased = purchase_completed[i]

  print(f"\nCustomer: {customer}")
  print(f"  Internet Speed: {internet_spd}")
  print(f"  Purchased Completed: {purchased}")

  if purchased == 'no' and internet_spd < 3:
    print("  Diagnosis: Customer abandoned purchase, and slow internet might be a contributing factor.")
  elif purchased == 'no' and internet_spd >= 3:
    print("  Diagnosis: Customer abandoned purchase, but internet speed was not a reported issue.")
  else:
    print("  Diagnosis: Purchase completed. Internet speed seems adequate.")

```

```
#Payment Diagnosis
print('\n--- Payment Diagnosis ---')
for i in range(len(Cust_names)):
  customer = Cust_names[i]
  payment_fail = payment_failure[i]
  purchased = purchase_completed[i]

  print(f"\nCustomer: {customer}")
  print(f"  Payment Failure Count: {payment_fail}")
  print(f"  Purchased Completed: {purchased}")

  if purchased == 'no' and payment_fail > 0:
    print("  Diagnosis: Customer abandoned purchase, and payment failures might be the reason.")
  elif purchased == 'no' and payment_fail == 0:
    print("  Diagnosis: Customer abandoned purchase, but payment was not a reported issue.")
  else:
    print("  Diagnosis: Purchase completed. Payment was successful.")
```

```
# Delivery Diagnosis
print('\n--- Delivery Diagnosis ---')
for i in range(len(Cust_names)):
  customer = Cust_names[i]
  delivery_charge = delivery_charges[i]
  purchased = purchase_completed[i]

  print(f"\nCustomer: {customer}")
  print(f"  Delivery Charges: {delivery_charge}")
  print(f"  Purchased Completed: {purchased}")

  if purchased == 'no' and delivery_charge > 30: # Assuming charges > 30 are high enough to deter
    print("  Diagnosis: Customer abandoned purchase, and high delivery charges might be a contributing factor.")
  elif purchased == 'no' and delivery_charge <= 30:
    print("  Diagnosis: Customer abandoned purchase, but delivery charges were not excessively high.")
  else:
    print("  Diagnosis: Purchase completed. Delivery charges seem acceptable.")
```

```
# ROOT CAUSE ANALYSIS
print("\n Root Cause Analysis")
for i in range (len(Cust_names)):
  if(purchase_completed[i])== "false":
    if internet_speed[i] < 2 :
      print(Cust_names[i],"abandoned cart due to slow internet")
    elif payment_failure[i]>0:
      print(Cust_names[i],"abandoned cart due to payment failure")
    elif delivery_charges[i]>40:
      print(Cust_names[i],"abandoned cart due to high delivery charges")
```
