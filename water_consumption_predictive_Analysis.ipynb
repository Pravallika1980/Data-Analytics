```python
import statistics
import matplotlib.pyplot as plt
import numpy as np

water_consumption=[10,20,30,40,50,55,60]
days=[1,2,3,4,5,6,7]
 #analysis
 #avergae water consumption
 #total usage in a day
 #water usage increased
 #sumof water increased daily
 #prediction
```

```python
#average water consumption
average_water_consumption=sum(water_consumption)/len(water_consumption)
print(average_water_consumption)
```

```python
 #total usage in a day
total_water_consumption=sum(water_consumption)
print(total_water_consumption)
```

```python
#water usage increased
water_usage_increased=water_consumption[2]-water_consumption[1]
print(water_usage_increased)
```

```python
#sumof water increased daily
sum_of_water_increased_daily=sum(water_consumption[1:])-sum(water_consumption[:-1])
print(sum_of_water_increased_daily)
```

```python
#prediction
prediction=sum_of_water_increased_daily+water_consumption[-1]
print(prediction)
if sum_of_water_increased_daily<average_water_consumption:
  print("water usage is decreased")
else:
  print("water usage is increased")
```
